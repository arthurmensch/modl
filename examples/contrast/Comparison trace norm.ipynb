```python
from pymongo import MongoClient
import pandas as pd
import seaborn as sns
```

```python
client = MongoClient()

db = client.amensch
runs = db.drago2

multi_dataset_res = runs.aggregate([{"$match": {"info.multi_predict_contrast_id": {'$in': [668]},
                                               }},
                                    {'$project': {'penalty': '$config.penalty',
                                                  'alpha': '$config.alpha',
                                                  'seed': '$config.seed',
                                                  'test_score': '$info.test_archi_score',
                                                  'train_score': '$info.train_archi_score',
                                                  "datasets": "$config.datasets"}}])

multi_dataset_res = list(multi_dataset_res)
df = pd.DataFrame(multi_dataset_res)
df['datasets'] = df['datasets'].map(lambda x: '_'.join(x) if isinstance(x, list) else x)
df = df.set_index(['datasets', 'alpha', 'penalty', 'seed'])
df.sort_index(inplace=True)

df_agg = df.groupby(level=['datasets', 'alpha', 'penalty']).aggregate(['mean', 'std'])['test_score']
argmax = df_agg.loc[df_agg['mean'].argmax()]
```

```python
df.loc[df_agg['mean'].argmax()]
```

```python
client = MongoClient()

db = client.amensch
runs = db.drago

multi_dataset_res = runs.aggregate([{"$match": {"info.multi_predict_contrast_id": {'$in': [2705]},
                                                "config.datasets": {"$in": ['hcp']}
                                               }},
                                    {'$project': {'penalty': '$config.penalty',
                                                  'alpha': '$config.alpha',
                                                  'beta': '$config.beta',
                                                  'latent_dim': '$config.latent_dim',
                                                  'dropout': '$config.dropout',
                                                  'activation': '$config.activation',
                                                  'seed': '$config.seed',
                                                  'test_score': '$info.test_hcp_score',
                                                  'train_score': '$info.train_hcp_score',
                                                  "datasets": "$config.datasets"}}])

multi_dataset_res = list(multi_dataset_res)
df_latent = pd.DataFrame(multi_dataset_res)
df_latent['datasets'] = df_latent['datasets'].map(lambda x: '_'.join(x) if isinstance(x, list) else x)
df_latent = df_latent.set_index(['datasets', 'alpha', 'beta', 'latent_dim', 'activation', 'dropout', 'seed'])
df_latent.sort_index(inplace=True)

df_agg_latent = df_latent.groupby(level=['datasets', 'alpha', 'beta', 'latent_dim',
                           'activation', 'dropout']).aggregate(['mean', 'std'])['test_score']

argmax = df_agg_latent.loc[df_agg_latent['mean'].argmax()]
```

```python
simple_dataset_res = runs.aggregate([{"$match": {"info.multi_predict_contrast_id": {'$in': [1]},
                                                 "config.datasets": {"$not": {"$all": ['archi','hcp']}},
                                                }},
                                    {'$project': {'penalty': '$config.penalty',
                                                  'alpha': '$config.alpha',
                                                  'seed': '$config.seed',
                                                  "datasets": "$config.datasets",
                                                  'test_score': '$info.test_archi_score',
                                                  'train_score': '$info.train_archi_score',
                                                  'test_hcp_score': '$info.test_hcp_score',
                                                  'train_hcp_score': '$info.train_hcp_score'}}])
simple_dataset_res = list(simple_dataset_res)
df_s = pd.DataFrame(simple_dataset_res)    
df_s = df_s.set_index(['alpha', 'penalty', 'seed'])
df_s.sort_index(inplace=True)

df_s_agg = df_s.groupby(level=['alpha', 'penalty',]).aggregate(['mean', 'std'])['test_score']

argmax = df_s_agg.loc[df_s_agg['mean'].argmax()]
```
